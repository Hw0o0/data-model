```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rc
```

# CSV에서 데이터 읽기

```python
# csv에서 데이터 읽기
csv_file_path = '/Users/gangjeongmu/Desktop/data-model/data-model/A회사_보습제_매출데이터_v1.csv'
data = pd.read_csv(csv_file_path)

data = pd.read_csv(csv_file_path, skiprows=1)

data.columns = ['카테고리', '9월', '10월', '11월', '12월', '1월', '2월']

data.replace({'\$': '', ',': ''}, regex=True, inplace=True)
data.iloc[:, 1:] = data.iloc[:, 1:].astype(float)
```

```python
# 수정된 데이터프레임 확인
data
```

# 데이터 시각화

```python
# 한글 출력을 위한 폰트 설정
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 그래프를 그리기 위해 전치
data_t = data.set_index('카테고리').T
```

```python
# 총매출, 목표매출 그래프
plt.plot(data_t.index, data_t['총매출'], marker='o', label='총매출')
plt.plot(data_t.index, data_t['목표매출'], marker='x', linestyle='--', label='목표 매출')
plt.title('총 매출, 목표 매출 비교')
plt.xlabel('월')
plt.ylabel('매출')
plt.legend()
plt.show()
```

```python
# 광고비용 그래프
plt.plot(data_t.index, data_t['광고비용'], marker='o', linestyle='-', color='purple', label='광고 비용')

# 소셜네트워크비용 그래프
plt.plot(data_t.index, data_t['소셜네트워크비용'], marker='s', linestyle='-', color='green', label='소셜 네트워크 비용')

# 그래프 제목과 레이블 추가
plt.title('시간별 광고, 소셜네트워크비용 비교')
plt.xlabel('월')
plt.ylabel('비용')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1온스 별 단가 그래프
plt.plot(data_t.index, data_t['1온스별단가'], marker='o', linestyle='-', label='1온스 별 단가')
plt.title('시간별 1온스 별 단가 변화')
plt.xlabel('월')
plt.ylabel('단가')
plt.legend()
plt.show()
```
