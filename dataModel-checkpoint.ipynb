```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv(r"C:\Users\UIT801-\Downloads\A회사_보습제_매출데이터_v1.csv", encoding='cp949')

# 데이터 확인
print(data)

# 월별 매출과 비용 데이터 추출
월 = ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb']
광고비용 = pd.to_numeric(data.loc[2][1:].str.replace(',', '').str.replace('$', ''), errors='coerce').dropna()
소셜네트워크비용 = pd.to_numeric(data.loc[3][1:].str.replace(',', '').str.replace('$', ''), errors='coerce').dropna()

# 시각화
plt.figure(figsize=(10, 6))

plt.plot(월, 광고비용, marker='o', label='광고비용')
plt.plot(월, 소셜네트워크비용, marker='o', label='소셜네트워크비용')

plt.title('월별 매출과 비용 변화')
plt.xlabel('월')
plt.ylabel('금액')
plt.legend()
plt.grid(True)

# y축 범위 설정
plt.ylim(0)  # 이전 코드에서 수정이 필요하면 값 변경 가능

plt.show()
```
